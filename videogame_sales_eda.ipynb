```python
#removing outliers
import numpy as np
x=["NA_Sales","EU_Sales","JP_Sales","Other_Sales"]
for i in x:
  data=df[i]
  q1=np.quantile(data,0.25)
  q2=np.quantile(data,0.50)
  q3=np.quantile(data,0.75)
  iqr=q3-q1
  lower_bound=q1-(2*iqr)
  upper_bound=q3+(2*iqr)
  outliers=[j for j in data if j<lower_bound or j>upper_bound]
  print(len(outliers))
  df=df[(df[i]>= lower_bound) & (df[i]<=upper_bound)]
```

#**DIMENTIONALITY REDUCTION**

Reduce variables in training dataset


```python

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x = [ "Platform", "Year", "Genre", "Publisher", "NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]
a = sc.fit_transform(df[x])
pca = PCA(n_components=2)
pca_r = pca.fit_transform(a)
df["PCA"] = pca_r[:, 0]  # Adds 'PCA' column to df1
df["PCA2"] = pca_r[:, 1]
```

To take all contents and create new two features to check

```python
#model evaluation
from sklearn.model_selection import train_test_split
y=df["Global_Sales"]
x=df.drop(columns=["Global_Sales"])
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,shuffle=True)
print(f"x_train shape: {x_train.shape}")
print(f"x_test shape: {x_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
print(x_train)
y_train
y_test
```

#**INITIAL MODELING AND HYPOTHESIS**

```python
from sklearn.svm import SVR
svm_model = SVR()
#fit the model to train data
svm_model.fit(x_train, y_train)
#predict the target value of the test data
svm_pred = svm_model.predict(x_test)
print("SVM_Prediction",svm_pred) #y_test predict the value
```

predict the target value of test dataset

#**DOCUMENTATION OF REPORTING**

```python
#calculate error
from sklearn.metrics import mean_absolute_error
svm_mae=mean_absolute_error(y_test,svm_pred)
print("Mean Absolute Error:",svm_mae)
```

Mean squared error of support vector machine is 0.049349667789456424

```python
from sklearn.metrics import mean_squared_error
import math #import the math module
svm_mse=mean_squared_error(y_test,svm_pred)
svm_rmse=math.sqrt(svm_mse) #use math.sqrt to calculate square root
print("Root Mean Squared Error:",svm_rmse)
print("Mean Squared Error:",svm_mse)
```

root means squared error of the support vector machine is 0.05554536881264972

```python
from sklearn.metrics import r2_score
r=r2_score(y_test,svm_pred)
print("R2 Score:",r)
```

statistical metrics used in machine learning to evaluate a regression model's quality is  0.6859732236586282
